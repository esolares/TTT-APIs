## Google TTS

```python
import speech_recognition as sr
from gtts import gTTS
from IPython.display import Audio
```

```python
def speak(text):
    #tts = gTTS(text = text, lang = 'en', tld = 'com.au')
    tts = gTTS(text = text, lang = 'en')
    filename = 'voice.mp3'
    tts.save(filename)
    display(Audio(filename, autoplay=True))
```

```python
speak('Principal Component Analysis relies on Eigen Vectors and Values as well as Projections and Dimensional Reduction. Using PCA we can construct a co-variance matrix')
```

```python
speak('Hello! Ara baila bonita')
```
